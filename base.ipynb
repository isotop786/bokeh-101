```python

output_file("student.html")

# create a figure object
f = figure()

f.line(bachelor['Year'],bachelor['Engineering'])
f.line(bachelor['Year'],bachelor['Computer Science'],name="computer",line_color="red")
f.line(bachelor['Year'],bachelor['Business'],line_color="blue")
f.line(bachelor['Year'], bachelor['Art and Performance'],line_color="#555",name="art")
f.line(bachelor['Year'], bachelor['English'],line_color="purple")
    
show(f)
```

```python


p = figure(plot_width=1000,plot_height=700)
output_file("gradutes.html")


p.line(bachelor['Year'],bachelor['Engineering'])


p.title.text = "Graduation Data"
p.title.text_color = "Gray"
p.title.text_font = "times"
p.title.text_font_style = "bold"
p.xaxis.axis_label="Year of Graduation"
p.yaxis.axis_label="Number of Graduate Students"




show(p)
```

```python
wd = pandas.read_excel('verlegenhuken.xlsx')
```

```python
# producing graph based on weather data\


w = figure(plot_width=1000, plot_height=700)
w.xaxis.axis_label = 'Temperature (C)'
w.yaxis.axis_label = 'Pressure (hPa) '
w.title.text = "Temperature and Air Pressure"
w.title.text_color = "Gray"
w.title.text_font="Arial"
w.title.text_font_style= "bold"


w.dot(wd['Temperature'],wd['Pressure'])

output_file('weather.html')

show(w)

```

```python
from bokeh.plotting import figure, output_file, show 
import pandas 

adbe = pandas.read_csv('adbe.csv',parse_dates=['Date'])


p = figure(x_axis_type="datetime")
p.width = 1400
p.height =600
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"
output_file('adbe.html')

p.line(adbe['Date'],adbe['Close'],line_color="Orange")

show(p)

```
